### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Multiple Plot Types in 2x2 Layout', fontsize=16)

# Subplot 1: Line Chart (top-left)
x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)
axs[0, 0].plot(x_line, y_line, 'b-', linewidth=2)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Sin(x)')
axs[0, 0].grid(True, linestyle='--', alpha=0.7)

# Subplot 2: Bar Chart (top-right)
categories = ['A', 'B', 'C', 'D', 'E']
values = [12, 9, 15, 7, 10]
bar_colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
axs[0, 1].bar(categories, values, color=bar_colors)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Subplot 3: Scatter Plot (bottom-left)
x_scatter = np.random.rand(30) * 10
y_scatter = np.random.rand(30) * 10
sizes = np.random.rand(30) * 100 + 50
axs[1, 0].scatter(x_scatter, y_scatter, s=sizes, c=x_scatter, cmap='viridis', alpha=0.7)
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')
axs[1, 0].grid(True, linestyle='--', alpha=0.7)

# Subplot 4: Pie Chart (bottom-right)
labels = ['Group A', 'Group B', 'Group C', 'Group D']
sizes_pie = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # explode the 2nd slice
colors_pie = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Remove axis for pie chart subplot
axs[1, 1].axis('equal')
axs[1, 1].pie(sizes_pie, explode=explode, labels=labels, colors=colors_pie,
              autopct='%1.1f%%', shadow=True, startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(top=0.9)  # Adjust for the main title

# Show the plot
plt.show()
```
